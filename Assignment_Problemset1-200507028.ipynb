# Question 3 (C#)

```python
using System;
class Triangle
{
    static void Main(string[] args)
    {
        //getting side 1 of the triangle 
        Console.Write("Enter Side 1: ");
        double side1 = Convert.ToDouble(Console.ReadLine());

        //getting side 2 of the triangle 
        Console.Write("Enter Side 2: ");
        double side2 = Convert.ToDouble(Console.ReadLine());


        //getting side 3 of the triangle 
        Console.Write("Enter Side 3: ");
        double side3 = Convert.ToDouble(Console.ReadLine());

        //Calculation of S
        double S = (side1 + side2 + side3) / 2;

        //Calculation of area using herons formula 
        double area = Math.Sqrt(S * (S - side1) * (S - side2) * (S - side3));
        Console.WriteLine("Area of the trianle of the sides " + side1 + " " + side2 + " " + side3 + " " + "is : " + area);

    }
}
```

# Question 4 (C#)

```python
using System.Linq;
namespace MyApp // Note: actual namespace depends on the project name.
{
    internal class Program
    {
        static void Main(string[] args)
        {
            



            Console.WriteLine("Define Array Size? ");
            int number = Convert.ToInt32(Console.ReadLine());

            Console.WriteLine("Enter numbers:\n");
            int[] arr = new int[number];
            List<int> oddNumbers = new List<int>();
            List<int> evenNumbers = new List<int>();


            for (int i = 0; i < number; i++)
            {
                arr[i] = Convert.ToInt32(Console.ReadLine());
            }
            for (int i = 0; i < arr.Length; i++)
            {
                Console.WriteLine("Array Index: " + i + " AND Array Item: " + arr[i].ToString());
            }
          


            for (int i = 0; i < number; i++)
            {
                if (arr[i]%2 == 0) 
                {
                    evenNumbers.Add(arr[i]);
                }
                else
                {
                    oddNumbers.Add(arr[i]);
                }
            }
            Console.WriteLine("Even elements in the array are :");
            evenNumbers.ForEach(Console.WriteLine);

            Console.WriteLine("\nEven elements in the array are :");
            oddNumbers.ForEach(Console.WriteLine);
        }
    }
}
```

# Question 5 (C#)

```python
class GFG
{

    //creating function
    static bool inside(double x1, double y1, double x2,
                          double y2, double x, double y)
    {
        if (x > x1 && x < x2 &&
            y > y1 && y < y2)
            return true;

        return false;
    }

    // Driver code
    public static void Main()
    {

        // lower-left x
        Console.WriteLine("enter x for lower left conrner");
        double x1 = Convert.ToDouble(Console.ReadLine());

        // lower-left y
        Console.WriteLine("enter y for lower left conrner");
        double y1 = Convert.ToDouble(Console.ReadLine());


        // upper-right x 
        Console.WriteLine("enter x for upper right conrner");
        double x2 = Convert.ToDouble(Console.ReadLine());

        // upper-right y
        Console.WriteLine("enter y for upper right conrner");
        double y2 = Convert.ToDouble(Console.ReadLine());

        // given point x to check 
        Console.WriteLine("enter x for checking");
        double x = Convert.ToDouble(Console.ReadLine());

        // given point y to check 
        Console.WriteLine("enter y for checking");
        double y = Convert.ToDouble(Console.ReadLine());



        // function call
        if (inside(x1, y1, x2, y2, x, y))
            Console.Write("True");
        else
            Console.Write("False");
    }
}
```

```python
def frequencies(string):
    a = []
    only_alpha = ""
    for char in string:
        if ord(char) >= 65 and ord(char) <= 90:
            only_alpha += char

        elif ord(char) >= 97 and ord(char) <= 122:
                only_alpha += char
    
    lowered_alpha = only_alpha.lower()


    Alphabet = [
    'a','b','c','d','e','f','g','h','i','j','k','l','m',
    'n','o','p','q','r','s','t','u','v','w','x','y','z']

    for i in range(0,len(Alphabet)):
    
    
         
        b = lowered_alpha.count(Alphabet[i])
        a.append(b)   

    return a
```
